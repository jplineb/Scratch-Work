# Load data into Pandas Dataframe

We can load data into pandas using pd.read_csv('filename.csv') using .csv, .txt, .excel, etc format

```python
loan_data = pd.read_csv('./data/loan_data.csv') # use the read function to create a new dataframe called "loan_data"
loan_data.head() # displays the dataframe in jupyter notebooks
```

# Information about dataframes
Python can be a very object-oriented language depending on the approaches of certain packages. For Pandas, a dataframe is an object; it contains not just the data, but also sub methods or 'functions'.

```python
# Example 1: We can access the mean of every column by calling the mean method
loan_data.mean(numeric_only=True)
```

```python
# Example 2: Getting stastical information about a dataframe
loan_data.describe()
```

```python
# Example 3: ploting a specific column
loan_data.installment.plot()
```

# Accessing data in dataframes 

```python
# we can get certain rows of a dataframe by calling the .iloc function
loan_data.iloc[0:5] # get rows 0-4
```

```python
# Or we can also get columns by interger
# in this case we will get the installment column
loan_data.iloc[:,3] # Explanation: From loan_data "return all the data from" .iloc[all data in the series, column number]
```

```python
# or by using the loc function we can reference the column by name
loan_data.loc[:,'installment']
```

## We can also get a column name by treating the column name as a method or indexing it like a dictionary

```python
# "method" method
loan_data.installment
```

```python
# "key value pair" method
loan_data['installment']
```

# Data preprocessing tips

## Checking for null value

Most important first step to take with large datasets is check for null values. This will cause headaches later on if you don't. There are many ways to do so which will be explored

```python
# This will return the number of null values in each column
loan_data.isnull().sum()
```

```python
# To return a dataframe with rows that have null values
loan_data[loan_data.isnull().any(axis=1)]
# This will return an empty dataframe because there are not any empty values
```

## Dealing with null values

```python
# Dropping all rows with a column that has an empty value
loan_data.dropna(how='any',axis=0)
# Because we don't have any null values, no rows are dropped
```
